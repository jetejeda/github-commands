## Branches 

They are an exact copy of the main branch in which we can upload changes without the risk of cracking the main branch. Is a new timeline for new commits.

The main idea of a branch is to have a safe testing environment that can be merged with other branches including the main branch.

### Merges
Is when we want to joint a branch into another one. Git will try to do this automatically. 

There are three different scenarios for merges:

1. Fast-forward: is simple to merge the branches, git can successfully do it automatically.
2. Automatic Merge: The main branch has a change that the branch doesn't know about, but the merge can still be completed automatically.
3. Manual Merge: When git cannot automatically merge the branches. It's caused because the two branches that we want to merge have changes in the same code. We have to manually decide what code will be included, finally, git will create a new commit (Merge commit).

If we return to the main branch which is outdated we are going to go back to the last commit that the main branch has, this means that we are going to miss the changes that we did in the new branch.

## Semantic version of tags

When we have reached a milestone with our code.

git tag -a v1.0.0 -m "version 1.0.0 ready" # Will create a release tag.

1. Syntax with release versions: for mayor changes related tu functionality we have to modify the first digit 1.0.0 -> 2.0.0. It's a major version
2. When we add new functionality or to the packages of the app. 2.0.0 -> 2.1.0
3. For bug fixing we have to modify the last digit 2.0.0 -> 2.0.1




With tags we can do time traveling.

# Git stash

It's a secure version of the branch that we are working on. We can include all the changes that we have and also all the untracked files for git.

We will have all the code in the stash and we can restore it.

It's not recommended to have multiple stashes.

We only have to use it when we are in a hurry to deploy to production. When we are still working in new functionalities that could crack production and we don't want to loose them. If we are working with branches we should not need to use the stash, we should just move to the other branch and do the modifications there.

It will save everything that is in the working directory, even files that git don't keep track of them.

git stash #Create a new instance for the stash.



WIP -> Work In Progress.


After the stash it will delete all the changes after the last commit, every change is now safe at the stash. 

git stash list # Display the stack of stashes.

We could have multiple stashes but it's recommended to have as few as possible.

Will display all the stashes that we have.



After all the changes had been added to the stash, we could modify the branches and continue our normal work doing commits an so on.



Git will not remember us about the stash.

git stash pop #It will return all the changes that are in the latest stash at that time. It will also delete the stash instance.

# Git rebase

git rebase branchName #To do a rebase from the specified branch to the current one.



## Use case

When we have two branches and both of them had received distinct commits.

The rebase will help us to update the working branch with the latest changes/commits from the main branch and then reincorporate the changes that we had in the other branch.

Do not use rebase or name changing for commits if the history is already in a remote repository.

## Interactive git rebase

git rebase -i HEAD~numberOfCommits #Enter the interactive mode for git rebase.



In this example we have a branch called "rama-misiones-completadas" and the main branch, both have been updated but never merged, the commits that are in the main branch should be part of the other one.

When performing a rebase, we should be in the branch in which we want to receive all the changes. In this case we want the branch called "rama-misiones-completadas" to receive all the changes from the main branch, so we must be in that branch.

## Rebase reword

Re-type the commit message from any of all the commits that we specify in the interactive rebase.

It will only update the name, not who did it.
1. git rebase -i HEAD~numberOfCommits
2. We will see the number of commits that we specified in the previous step.
3. We can modify the name by adding the r or reword keyword in the specific commit.
4. Save changes using ESC : wq!
5. In the next view modify the commit or simply close the file.
6. Repeat step 5 for all the commits that were affected.
