<h1>basic commands</h1>

For git a project is equal to a repository.

There are two types of repositories (repo's):
1. Central repo: Where the complete team works in a single server.



2. Distributed repo: This is how git works. Every member of the team has an exact local copy of the repository. There is also a central repository in which every member push it's changes and receives the ones made by their teammates. It's more secure because we have a copy from the repository.





<h1>Commits</h1>

They are like snapshots of the code.

<h1>basic commands</h1>

git <b>--</b>version

The -- means that after it, we will find a complete word.
git commit <b>-</b>am

The single - means that after it, we will find a trimmed word.

git --version #tells us the git version of our computer.

git help #To get the guide helper.

git help keyword #we can substitute the keyword for any of git's keywords to get specific information about that command.

<b>To exit the helper we just have to press the Q key</b>

<h1>git config</h1>

Used to get and set repository or global options.

We could use --global to set parameters globally. It will apply the configurations to the OS user.
    
    Examples: 
        git config --global user.name "My name"
        git config --global user.email "MyEmail@dom.com"
    
    git config --global -e #Used to list all the global configurations that we have
        It will display all the variables, we can modify their values va pressing the a keyword. After we have modified the values we have to press esc or :, wq! to save the changes wq! stands for write quit ! (for close)
        
git config --global init.defaultBranch branchName #To specify that every time we create a new repository using git init we will have that specific default branch.

<h1>Creating our first repository</h1>

git init # Used to initialize a new repository

git status #Used to see info about commits, the currently branch and all the untracked files.



git add fileName fileName2...#To add a new file or list of files to the next commit.

git add . #Adds all the changes that we have done to the next commit

git reset fileName # We can use this commit to remove an specific file from the next commit

git commit -m "information about the commit"

git log #We can use this command to see all the commits that we have done

git checkout -- . #Use this to reverse all changes to the last commit.

git branch #Get all branches, it will display a "*" BranchName for the currently branch

<p>A branch is the place in which we are working, we can have multiple branches.
We can think of a repository as a tree in which we can have multiple branches, and every branch can return to the main branch.

In the main branch we should only have all the things that are officially going to production environments.
</p>

git branch -m branchName newName #Change a branch name

git config --global init.defaultBranch branchName #Set globally the main branch name.

# Shortcut to commit files that are already in the repository:

git commit -am "commit info" #will stage the changes and add them in the commit

# Different ways to add files to the staging phase

It's a best practice that the files we are going to commit are related by an specific functionality only.

git add *.fileExtension #Wildcard to add all the files from an specific file extension, but only from the main folder.

git add directory/*.js #Wildcard to add all the files from an specified directory that matches the fileExtension.

<p>If we create an empty folder, git will not keep track of it until it has some content.

In order to keep track of an empty folder we must add a .gitkeep file
</p>

git add folder/ #Will add all files and sub-folders that are contained in the folder that we specified.

# shortcut commands:

# Working tree

It's a tool that git uses to keep track of the changes from the last commit to the current version.

git diff #Show modifications per file. It will only show files that are modified but are not included in the stage by the time.

git diff --staged #Show modifications from the last commit to the files that are in the stage.





# Update a commit message



<b>git commit --amend -m "fundamentals updated"</b>



# Revert commits

git reset --soft HEAD^number#To reverse the current commit but maintain the changes. HEAD points to the last commit. We could use the HEAD keyword or the hash of a commit. The number after the ^ symbol is used to point to previous commits than main.

For example, the current command will return the "Global commands and shortcuts" commit:

git reset --soft HEAD^2



<b><center>It's not recommended to use the git reset, it can lead to future conflicts with the repository</center></b>

git reset --mixed commitHash #It will remove everything from the stage but all the changes remain ready to be added.



git reset --hard #It will reverse the current commit and discard all the changes.
